```python
# variable list dictionaries conditions and loops
# student dictionary of 5 students name,study_hours,attendence and marks
# student name as input and based on their marks you have to consider a print statement for the performance
# marks is more than 75 then performance is strong
# marks is equal or greater than 40 performance should need improvement
# marks less than 20 then print fail repeat class

# attendence 75 warning low attendence
# study hours less than 4 print should come warning needs more study time


students = {
    "Rahul": {"study_hours": 5, "attendance": 80, "marks": 85},
    "Priya": {"study_hours": 3, "attendance": 70, "marks": 65},
    "Aman": {"study_hours": 2, "attendance": 60, "marks": 15},
    "Neha": {"study_hours": 4, "attendance": 90, "marks": 78},
    "Riya": {"study_hours": 3, "attendance": 75, "marks": 35}
}

name = input("Enter student name: ")

if name in students:
    student = students[name]

    marks = student["marks"]
    attendance = student["attendance"]
    study_hours = student["study_hours"]

    if marks > 75:
        print("Performance: Strong")
    elif marks >= 40:
        print("Performance: Need Improvement")
    elif marks < 20:
        print("Performance: Fail, Repeat Class")
    else:
        print("Performance: Below Average")


    if attendance < 75:
        print("Warning: Low Attendance")


    if study_hours < 4:
        print("Warning: Needs More Study Time")
```
